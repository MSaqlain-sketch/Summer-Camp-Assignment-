8. A school has following rules for grading system:

a. Below 25 - F

b. 25 to 45 - E

c. 45 to 50 - D

d. 50 to 60 - C

e. 60 to 80 - B

f. Above 80 - A

Ask user to enter marks and print the corresponding grade.

```python
mark=int(input("Enter your marks :"))
if mark>80:
    print("A Grade")
elif mark>60:
    print("B Grade")
elif mark>50:
    print("C Grade")
elif mark>45:
    print("D Grade")
elif mark>25:
    print("E Grade")
else:
    print("F Grade")        
```
